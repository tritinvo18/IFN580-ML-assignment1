This part is only check the basic info of the dataset, doesn't include cleaning or preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read dataset
data_path = 'Asm1_dataset26.csv'
df = pd.read_csv(data_path)

print(f'Dataset shape: {df.shape}')
df.head()
```

```python
# column and type
df.info()
```

```python
# number type values
df.describe()
```

```python
# include all the variables
df.describe(include='all')
```

```python
# missing data count
missing_count = df.isnull().sum().sort_values(ascending=False)
missing_ratio = (df.isnull().mean() * 100).sort_values(ascending=False)

missing_summary = pd.DataFrame({
    'missing_count': missing_count,
    'missing_ratio_%': missing_ratio
})

missing_summary[missing_summary['missing_count'] > 0].head(30)
```

```python
# the number of repeating column
duplicate_rows = df.duplicated().sum()
print(f'Duplicate rows: {duplicate_rows}')
```

```python
# number type variables distribution
numeric_cols = df.select_dtypes(include=[np.number]).columns

if len(numeric_cols) > 0:
    df[numeric_cols].hist(figsize=(14, 10), bins=30)
    plt.tight_layout()
    plt.show()
else:
    print('No numeric columns found.')
```

```python
# heatmap
if len(numeric_cols) > 1:
    corr = df[numeric_cols].corr(numeric_only=True)
    plt.figure(figsize=(12, 8))
    sns.heatmap(corr, cmap='coolwarm', center=0)
    plt.title('Correlation Heatmap (Numeric Features)')
    plt.show()
else:
    print('Not enough numeric columns for correlation heatmap.')
```
